### Descrição da tabela:
● Entity: O nome da entidade associado ao usuário.
● Username: O nome de usuário associado ao post.
● Medium: A rede social (por exemplo, facebook, instagram, tiktok).
● Medium ID: O identificador único da mídia.
● Media Type: O formato da mídia (por exemplo, texto, imagem, vídeo).
● ID: O identificador único da entrada.
● Date: A data da publicação da mídia.
● Variety: A variedade de mídia (por exemplo, post, ig_reel, yt_shorts).
● Engagement: O número total de interações (curtidas, comentários,
compartilhamentos, etc.).
● Likes, Reactions, +1’s: O número de curtidas e reações.
● Comments, Replies: O número de comentários e respostas.
● Shares: O número de compartilhamentos.
● Reposts, Retweets: O número de republicações e retweets.
● Reach: O alcance total da mídia.
● Impressions: O número total de vezes que a mídia foi exibida.
● Estimated Impressions: As impressões estimadas.
● Paid Impressions: As impressões pagas.
● Paid Video Views: As visualizações de vídeos pagos.
● Views: O número total de visualizações.
● Avg. Time Watched (sec): O tempo médio assistido em segundos.
● Total Time Watched (sec): O tempo total assistido em segundos.
● Post Valuation in USD: A avaliação da postagem em dólares americanos.
● Follower Count: O número de seguidores do usuário.

```python
def usuario_x_tempo_engajamento(dataFrame, nomeUsuario):

    dataFrame = dataFrame[dataFrame["Username"]==nomeUsuario]
    dataFrame = dataFrame[["Username","Date","Engagement"]].drop("Username",axis=1)
    dataFrame = dataFrame.set_index("Date").sort_values(by="Date",ascending=True)

    delta_Engagement = [0]
    valores_Engajamento_sequencia = list(dataFrame['Engagement'])

    contador = 1
    while contador < len(valores_Engajamento_sequencia):

        dif = valores_Engajamento_sequencia[contador] - valores_Engajamento_sequencia[contador-1]
        delta_Engagement.append(dif)

        contador = contador + 1

    dataFrame['delta_Engagement'] = delta_Engagement
    
    return dataFrame

```

## Função para análise gráfica:

```python
def plotagem_tempo_x_metrica(dataFrame, nomeMetrica, comum_ou_log):
    
    import matplotlib.pyplot as plt
    import random
    
    plt.figure(figsize=(25,6))
    
    plt.xlabel("Dias")
    plt.ylabel(nomeMetrica)
    
    plt.title(f"Tempo x {nomeMetrica}")
    
    lista_cores = ["black","dimgray","gray","darkgray","silver","lightgray"]
    
    if comum_ou_log == "comum":
        corrector = 10000
        
        plt.plot(dataFrame.index, dataFrame[nomeMetrica]/corrector, color=random.choice(lista_cores))
        
        return f"{nomeMetrica} foi dividido por {corrector} para facilitar a visualização"
        
    elif comum_ou_log == "log":
        plt.yscale("log")
        plt.plot(dataFrame.index, dataFrame[nomeMetrica], color=random.choice(lista_cores))
    
        return f"{nomeMetrica} foi transformado em logaritmo para facilitar a visualização"
```

```python
def scatter_tempo_x_metrica(dataFrame, nomeMetrica, comum_ou_log):
    
    import matplotlib.pyplot as plt
    import random
    
    plt.figure(figsize=(25,6))
    
    plt.xlabel("Dias")
    plt.ylabel(nomeMetrica)
    
    lista_cores = ["black","dimgray","gray","darkgray","silver","lightgray"]
    
    plt.title(f"Tempo x {nomeMetrica}")
    
    if comum_ou_log == "comum":
        corrector = 10000
        
        plt.scatter(dataFrame.index, dataFrame[nomeMetrica]/corrector, color=random.choice(lista_cores))
        
        return f"{nomeMetrica} foi dividido por {corrector} para facilitar a visualização"
        
    elif comum_ou_log == "log":
        plt.yscale("log")
        plt.scatter(dataFrame.index, dataFrame[nomeMetrica], color=random.choice(lista_cores))
    
        return f"{nomeMetrica} foi transformado em logaritmo para facilitar a visualização"
```
